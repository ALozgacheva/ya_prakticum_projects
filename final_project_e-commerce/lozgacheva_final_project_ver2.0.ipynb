Для дальнейшего исследования необходимо `выделить` из дат `месяц`, `год`, `день недели` - добавить соответствующую информацию в столбец `month`, `year`, `day` соответственно. К дню недели прибавляем +1 для удобства, так у нас получится пн -1, вс-7.Сделаем это:

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['day'] = (df['date'].dt.dayofweek) + 1
```

Чтобы проверить `сезонность` - добавим к таблице столбец `season` к датафрейму. Напишем функцию `season`, принимающую номер месяца `от 1 до 12` из столбца с месяцем и возвращающуб время года, к которому этот месяц относится:

```python
def season(month):
    if  3 <= month <= 5:
        return 'весна'
    elif 6 <= month <= 8:
        return 'лето'
    elif 9 <= month <= 11:
        return 'осень'
    else:
        return 'зима'
    
df['season'] = df['month'].apply(season)
df.sample(5)
```

Времена года опрелелены.

Для проверки гипотезы о продажах в выходые и будние дни разобьем дни недели. Напишем 2 функции: `weekday` и `weekend`. Первая будет возвращать 1, если день относится к будням, 0 - если нет. Вторая возвращает 1, если день недели - выходной день, 0 - если будний. 

```python
#функция weekday, принимающая номер дня (от 1 до 5) и возвращающая будни
def weekday(day):
    if  1 <= day <= 5:
        return 1
    else:
        return 0
    
#функция weekend, принимающая номер дня (от 6 до 7) и возвращающая выходные
def weekend(day):
    if  6 <= day <= 7:
        return 1
    else:
        return 0
```

Добавляем по итогам функций 2 столбца в наш датафрейм:

```python
df['weekday'] = df['day'].apply(weekday)
df['weekend'] = df['day'].apply(weekend)
df.sample()
```

Также приведем к нижнему регистру значения в столбце с наименованием продукта:

```python
df['product'] = df['product'].str.lower()
```

### Число заказов по пользователям

Найдем `число заказов по пользователям` — создаем датафрейм c двумя столбцами:  `customer_id`, `order_id`. Назовем датафрейм `orders_by_customers`.

```python
orders_by_customers = (
    df.groupby('customer_id', as_index=False)
    .agg({'order_id': 'nunique'})
)
orders_by_customers.columns = ['customer_id', 'orders_id']
orders_by_customers.sort_values(by='orders_id',ascending=False).head(10)
```

`Точечный график количества заказов по пользователям`:

```python
x_values = pd.Series(range(0,len(orders_by_customers['orders_id'])))
plt.scatter(x_values, orders_by_customers['orders_id'], color='lightseagreen')
plt.xticks(rotation=45)
plt.title('Точечный график количества заказов по пользователям ')
plt.show()
```

Видим выбросы - `126`, `35`, `17` заказов. 

```python
#считаем 95-й и 99-й выборочные перцентили количества заказов по 
#покупателелям методом np.percentile()
percentiles = np.percentile(orders_by_customers['orders_id'], [95, 99])
print('95-й перцентиль количества заказов на покупателя: {} заказа'.format(percentiles[0]))
print('99-й перцентиль количества заказов на покупателя: {} заказа'.format(percentiles[1]))
```

Не более `5%` пользователей совершили больше `2-ух` заказов. И не более `1%` клиентов – больше `3-ех` заказов. Разумно выбрать `3 заказа` на одного клиента за `нижнюю границу числа заказов`, и `отсеять аномальных клиентов` по ней. Для начала узнаем `количество` таких `аномальных пользователей`:

```python
anomaly_customers = pd.concat([orders_by_customers[orders_by_customers['orders_id']>3][
    'customer_id']], axis = 0)
print('Аномальных покупателей: {} человек'.format(len(anomaly_customers)))
```

Перед лемматизацией проводим `чистку данных`, чтобы в дальнейшем товары правильно распределялись по категориям. Для начала необходимо `удалить всю "шелуху" - символы`, которые встречаются в данных - сделать очистку данных, далее проведем `токенизацию` и `удаление стоп-слов`, далее уже делаем `лемматизацию`:

```python
#функция для очистки текста
def clean_text(text):
    # создаем регулярное выражение для удаления лишних символов
    regular = r'[\*+\#+\№\"\-+\+\=+\?+\&\^\.+\;\,+\>+\(\)\/+\:\\+]'
    # регулярное выражение для замены ссылки на "URL"
    regular_url = r'(http\S+)|(www\S+)|([\w\d]+www\S+)|([\w\d]+http\S+)'
    # удаляем лишние символы
    text = re.sub(regular, '', text)
    # заменяем ссылки на "URL"
    text = re.sub(regular_url, r'URL', text)
    # заменяем числа и цифры на ' NUM '
    text = re.sub(r'(\d+\s\d+)|(\d+)',' NUM ', text)
    # удаляем лишние пробелы
    text = re.sub(r'\s+', ' ', text)
    # возвращаем очищенные данные
    return text

# создаем список для хранения преобразованных данных 
processed_text = []
# загружаем стоп-слова для русского языка
stop_words = stopwords.words('russian')
# инициализируем лемматайзер 
lemmatizer = WordNetLemmatizer()

# для каждого продукта из столбца df_new['product']
for text in df_new['product']:
    # cleaning 
    text = clean_text(text)   
    # tokenization
    text = word_tokenize(text)       
    # удаление стоп-слов
    text = [word for word in text if word not in stop_words]     
    # лемматизация
    text = [lemmatizer.lemmatize(w) for w in text]
    
    # добавляем преобразованный текст в список processed_text
    processed_text.append(text)
    
# Сохраняем результат преобразования в новой колонке 'processed_product'
df_new['processed_product'] = processed_text
```

## Кластеризация, проверка гипотез

### Кластеризация

```python
#оставим только количественные данные
clusters = df_new.pivot_table(index = 'product', aggfunc = {'order_id': 'nunique',
                                                            'quantity': 'sum', 
                                                            'price': 'mean',
                                                            'weekday': 'sum', 
                                                            'weekend': 'sum'})
clusters.columns = ['orders', 'quantity', 'price', 'weekday', 'weekend']  
clusters.sample(5)
```

```python
# создаём объект класса scaler
scaler = StandardScaler() 
# обучаем нормализатор и преобразуем набор данных
x_sc = scaler.fit_transform(clusters)
```

Построим матрицу расстояний функцией `linkage()` на стандартизованной матрице признаков и нарисуем дендрограмму.

```python
linked = linkage(x_sc, method = 'ward') 
```

```python
plt.figure(figsize=(15, 10)) 
dendrogram(linked, orientation='top', show_leaf_counts=True, truncate_mode='lastp', p=30, leaf_rotation=90, 
           leaf_font_size=14)
plt.title("Дендрограмма клиентов интернет-магазина", fontsize=16)
plt.xlabel("Размер кластера")
plt.ylabel("Расстояние")
plt.show()
```

Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем сегменты покупателей:

```python
#задаём модель k_means с числом кластеров 2 и фиксируем значение random_state
km = KMeans(n_clusters=2, random_state=0)
#прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(x_sc)
```

```python
#сохраняем метки кластера в поле нашего датасета
clusters['cluster_km'] = labels
clusters
```

кластеризация выполнена. Добавляем к таблице `clusters` еще 2 столбца - выручка и средний чек. Соответственно `revenue`и `avg_check`, выведем статистику по кластерам:

```python
clusters['revenue'] = clusters['quantity'] * clusters['price']
clusters['avg_check'] = clusters['revenue'] / clusters['orders']
#статистика по кластерам
cluster_features = clusters.groupby('cluster_km').mean().reset_index()
numeric_columns = cluster_features.drop(columns=['cluster_km']).columns
cluster_features.style.format({i: '{:.2f}' for i in numeric_columns}) \
    .highlight_max(color='yellowgreen', subset=numeric_columns) \
    .highlight_min(color='coral', subset=numeric_columns)
cluster_features.T.style.background_gradient(cmap='rainbow', axis=1).set_precision(2)
```

По количеству заказов и средней суммарной выручке `кластер 1` выглядит лучше. При этом товаров продается больше в `кластере 0`, но средняя цена товаров в этом кластере ниже, чем в `кластере 1`. Средний чек при этом больше в `кластере 0`.

Исходя из статистики по кластерам можно сделать вывод, что в `кластере 0` недорогие товары, которые покупают чаще, в `кластере 1` товары более дорогие, но покупаются реже.  
Проверим, как часто эти товары покупают вместе. Создаем 2 списка товаров - `orders_0_list` и `orders_0_list`, соответственно в этих списках будут храниться товары 0 и 1 кластера, далее перебираем все товары в списке и в датафрейме `df_new` и в заранее созданный пустой список `orders_id_0` добавляем номера заказов, в которые попали товары из `кластера 0`, а в `orders_id_1` соответственно номера заказов, в которых есть товары из `кластера 1`, далее ищем пересеечнения (одинаковые товары в списках `orders_id_1` и `orders_id_0`), считаем количество таких заказов:

### Проверка гипотез

#### Гипотеза о среднем чеке

Формулируем нулевую и альтернативную гипотезы:

`H0` = Нет различий в среднем чеке по кластерам

`H1` = В среднем чеке кластеров есть статисттически значимое различие

```python
#генерируем две независимые выборки. 
cluster0 = clusters[clusters['cluster_km']==0]['avg_check']
cluster1 = clusters[clusters['cluster_km']==1]['avg_check']
```

Смотрим на соотношение дисперсий, и в случае различия более чем на 5 %, делаем вывод, что дисперсии не равны.

```python
print("Дисперсия 0-го кластера: {} ".format(np.var(cluster0, ddof=1)))
print("Дисперсия 1-го кластера: {} ".format(np.var(cluster1, ddof=1)))
print("Соотношение дисперсий: {} ".format((np.var(cluster1, ddof=1))/(np.var(cluster0, ddof=1))))
```

Чтобы рассчитать статистическую значимость различий в показателях между сегментами, применим критерий Манна-Уитни. 

```python
#проверка гипотезы
alpha = 0.05 # критический уровень статистической значимости 
results = stats.mannwhitneyu(cluster0, cluster1) 

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Смотрим медианнызначения:

```python
print("Медиана 0-го кластера: {} ".format(int(np.median(cluster0))))
print("Медиана 1-го кластера: {} ".format(int(np.median(cluster1))))
```

P-value = 0,083 значительно больше 0,05. Значит, причин отвергать нулевую гипотезу и считать, что в среднем чеке есть различия нет. То есть, нет статистически значимого различия в среднем чеке между между сегментами покупателей.

#### Гипотеза о продажах в выходные и будние дни

Формулируем нулевую и альтернативную гипотезы:

`H₀` = Средние продажи в выходные и будние дни одинаковые  
`H₁` = Средние продажи в выходные и будние дни разные

```python
revenue_weekdays = df_new[df_new['weekday']==1]['total']
revenue_weekends = df_new[df_new['weekend']==1]['total']
```

```python
print("Дисперсия продаж в будни: {} ".format(np.var(revenue_weekdays, ddof=1)))
print("Дисперсия продаж в выходные: {} ".format(np.var(revenue_weekends, ddof=1)))
print("Соотношение дисперсий: {} ".format((np.var(revenue_weekends, ddof=1))/(np.var(
    revenue_weekdays, ddof=1))))
```

```python
print("Стандартное отклонение объём продаж в будни: {} ".format(np.std(revenue_weekdays, ddof=1)))
print("Стандартное отклонение объём продаж в выходные: {} ".format(np.std(revenue_weekends, ddof=1)))
```

```python
print("Cреднее значение продаж в будни: {} ".format(int(np.mean(revenue_weekdays))))
print("Cреднее значение продаж в выходные: {} ".format(int(np.mean(revenue_weekends))))
```

```python
#проверка гипотезы
alpha = 0.05

results = stats.mannwhitneyu(revenue_weekdays, revenue_weekends) 

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

В этом случае есть вероятность получить случайно равенство продаж в выходные и будние дни.
